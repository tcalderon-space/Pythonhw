```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd 

from astropy import units as u
from astropy import constants as c

#### GAIA ARCHIVE WEBSITE ####
query_result = pd.read_csv('gaia_data_1.csv') 
```

```python
query_result
```

```python
#PART A
#this creates a new dataframe containing only those stars with valid luminosity values
query_result_filtered = query_result.dropna(subset=['lum_val'])
query_result_filtered
```

Number of stars reported are 6690. 

```python
#PART B
parallax = query_result_filtered['parallax']

parallaxerr = query_result_filtered['parallax_error']

ratio = (parallax/parallaxerr)

fig, ax = plt.subplots(figsize=(8,5), dpi=100)
ax.set_xlim(0,3000)
#plt.xticks(np.arange(0, 3000, 100))
ax.hist(ratio, bins = 250, color='#0099e6')

ax.set_ylabel('Signal/Noise')
ax.set_xlabel('Bins')

plt.show()
```

```python
#PART C
luminosities = query_result_filtered['lum_val']
temperatures = query_result_filtered['teff_val']

fig, ax = plt.subplots(figsize=(8,4), dpi=100)

ax.scatter(temperatures,luminosities, s=0.2, color='#ff0066')

ax.set_xlim(8000,3000)

ax.set_yscale('log')
#ax.set_xscale('log')

ax.set_ylabel(r'Luminosity [log($L_\odot$)]')
ax.set_xlabel(r'Effective Temperature [K]')

plt.show()
```

```python
#Part D
apparent_mag = query_result_filtered['phot_g_mean_mag']
BP_RP = query_result_filtered['bp_rp']
parallax = query_result_filtered['parallax']

d=1/(parallax)
abs_magnitude = apparent_mag - 5*np.log10(d/10)
fig, ax = plt.subplots(figsize=(6,4), dpi=100)

ax.scatter(BP_RP, abs_magnitude, s=0.1, color='#33cccc')
ax.set_ylim(25,18)

ax.set_ylabel(r'Absolute Magnitude')
ax.set_xlabel(r'BP-RP colour')

plt.show()
```
